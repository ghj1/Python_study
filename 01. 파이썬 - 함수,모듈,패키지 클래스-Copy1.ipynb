```python
def calc(n1, n2, op):
    if op == '+':
        return n1+n2
    if op == '-':
        return n1 - n2
    if op == '*':
        return n1 * n2
    if op == '/':
        return n1 / n2 

num1 = int(input('첫 번째 숫자: '))
num2 = int(input('두 번째 숫자: '))
op = input('연산 기호: ')
print('결과: %s' % calc(num1,num2,op))
```

```python
inStr = '''코스모스는 과거에 있었고, 현재에 있으며, 미래에 있을 그 모든 것이다.
우리는 코스모스의 일부다.
이것은 결코 시적 수사가 아니다.'''

# 방법1:
def  letterCount():
    global inStr
    countDic = {}
    countList=[]
    # 글자:빈도수 딕셔너리 만들기
    for char in inStr:
        if char >= '가' and char <='힣':
            if char in countDic:
                countDic[char] += 1
            else:
                countDic[char] = 1
                
    # 정렬하기
    # 방법 1
    countList = sorted(countDic.items(), key=lambda x: x[1], reverse=True)
    # 방법 2
    import operator
    countList = sorted(countDic.items(), key=operator.itemgetter(1), reverse=True)
    
    return countList

print(letterCount())

# 방법2
from collections import Counter
counter = Counter(inStr)  # 딕셔너리타입
print(counter)
# 가장 많이 나타난 2개를 출력한다
print(counter.most_common(n=2))
```

```python
# list 사용 

class Basket:
    
    def __init__(self,id):
        self.id = id
        self.items = []
        self.prices = []
        self.qtys = []
        self.totalprice = 0 
        self.itemCnt = 0
        
    def add(self, item, price, qty):
            self.items.append(item)
            self.prices.append(price)
            self.qtys.append(qty)
            self.totalprice += price * qty 
            self.itemCnt += 1
        
    def delete(self, item, qty):
        for i, v in enumerate(self.items):
            if item == self.items[i]:
                self.qtys[i] -= qty
                self.totalprice -= self.prices[i]*qty
                if self.qtys[i] == 0:
                    self.itemCnt -= 1
                    del(self.items[i])
                    del(self.prices[i])
                    del(self.qtys[i])
        
    def printitems(self):
        print(f'#{self.id}의 장바구니')
        print('-'*20)
        for i,v in enumerate(self.items):
            print(f'*{v}  {self.prices[i]}  {self.qtys[i]}')
        print(f'** 전체가격 = {self.totalprice} , 물건종류 {self.itemCnt}')
        print('-'*20)
        print()
        
        
basket1 = Basket("제인")
basket2 = Basket("대원")

basket1.add ('바나나',3000, 2)
basket1.add ('우유',2500, 2)
basket1.add ('라면',4000, 1)
basket1.delete ('우유',2)
basket2.add ('라면',4000, 2)

basket1.printitems()
basket2.printitems()
```

```python
# dict 사용 

class Basket:
    def __init__(self, id):
        self.id = id
        self.바구니={}
    def add(self, item, price, qty):
        self.바구니[item]=[price,qty]
    def delete(self, item, qty):
        self.바구니[item][1] -= qty
    def printitems(self):
        print(f'# {self.id}의 장바구니')
        print('-'*30)
        total = 0
        
        cnt = len(self.바구니)
        for key in self.바구니:
            print(f'* {key:10} {self.바구니[key][0]} {self.바구니[key][1]}')
            total += self.바구니[key][0] * self.바구니[key][1]
        print(f'** 전체가격 =  {total} 전체 수량 = {cnt}')
        
        
basket1 = Basket("제인")
basket2 = Basket("대원")
basket1.add('바나나',3000,2)
basket1.add('우유',2500,2)
basket1.add('라면',4000,1)
basket2.add('라면',4000, 2)
basket1.delete('우유',1)
basket1.printitems()
basket2.printitems()
```
